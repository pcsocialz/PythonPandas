```python
#Handling missing values in pandas
import pandas as pd
```

```python
#To parse date values as a timestamp use following command
df=pd.read_csv("/Users/purnachandra.dasari/Downloads/AIML-Labs-BLR/PythonTutorial-DS/weather_data_mv.csv",parse_dates=['day'])
# type(df.day[0]) : To check the type of the column
df.set_index('day',inplace=True)
df
```

```python
# If we want to fill all the na values with zero use the following command
newdf=df.fillna('0')
newdf
```

```python
#If we want to print na values by columns use the following command.
newdf=df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'no event'
})
newdf
```

```python
#Rather replacing with na with something else we can carry forward the previous day data to the na values
newdf_ffill=df.fillna(method='ffill') #ffill does the previous day data fill, #bfill does the next day data fill
newdf_ffill
newdf_bfill=df.fillna(method='bfill')
newdf_bfill
```

```python
# If we want to fill the values rather horizatally to do vertically use the following command
newdf_ffill=df.fillna(method='ffill',axis="columns") #ffill does the previous day data fill, #bfill does the next day data fill
newdf_ffill
newdf_bfill=df.fillna(method='bfill',axis="columns")
newdf_bfill
```

```python
df
```

```python
#To limit the forward fill to the certain number of missing rows use the following command
newdf_limit=df.fillna(method="ffill",limit=1)
newdf_limit
```

```python
newdf_int=df.interpolate(method="time")
#To get more details on interpolate use the below command
#http://pandas.pydata.org/pandas-docs/version/0.16.2/generated/pandas.DataFrame.interpolate.html
newdf_int
```

```python
#To drop all the na values use the following command (if row has at least one na)
newdf_dropna=df.dropna()
newdf_dropna
```

```python
#To drop all the na values use the following command (if row has all na)
newdf_dropna=df.dropna(how="all")
newdf_dropna
```

```python
#To drop all the na values use the following command (if at least one value exist keep the row) thresh=1, thresh=2 (at least two values should present)
newdf_dropna=df.dropna(thresh=2)
newdf_dropna
```
